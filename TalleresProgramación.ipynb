TALLER RSA

1.   El algoritmo de encriptación RSA nos ayuda a cubrir información que no queremos que alguien mas sepa. A diferencia de otros algoritmos, este no cuenta con el problema de ser simétrico, ya que no encripta con la misma llave que desencipta, evitando así que sea fácil de resolver al encontrar una llava por completo.

  El algoritmo RSA es asimétrico, lo que quiere decir que se encripta con una llave y se resuelve con otra. Una de estas llaves es pública, algo que puede sonar extraño ya que contradice el sentido común de seguridad, que es por lo que fue creado, sin embargo, esto no importa ya que la llave de desencriptación es privada, no se conoce, y la única forma de encontrarla es invirtiendo el algoritmo, lo cual es prácticamente imposible.

  Debido a esto, el algoritmo RSA es de los algoritmos de encriptación que más se usan en todo el mundo.
2.   El algoritmo se divide en 3:
  *   Generar la llave
  *   Encriptación
  *   Descifrado



  *   **GENERAR LA LLAVE**
    1.   Se escogen dos numeros primos $p$ y $q$
    2.   Hallamos un $n$, tal que $n = p\cdot q$

  ***Definicion:** Podemos decir que $a\equiv b\pmod n$, leído $a$ es congruente a $b$, sí y solo sí $n|(a-b)$. $n$ es el módulo.

    3.   $n$ será usado como el módulo de la llave pública y de la privada
  ***Definicion:** La función $\phi$ de Euler, $\phi(n)$, de un entero $p$ es la cantidad de números enteros positivos menores o iguales a $n$ y coprimos a $p$.
  ***Definicion:** Dos números $p$ y $q$ son coprimos, si el único divisor común es 1.
  ***Teorema:*** La función de Euler $\phi(n)$, de un número primo $p$, es $(p-1)$.
**Prueba** Si tenemos un número primo p, por definición sabemos que el único entero positivo menor o igual al primo $p$ que no es coprimo de $p$, es el mismo $p$. Por lo que $\phi(p) = p-1$.
  ***Teorema:** La función $\phi$ de Euler, $\phi(n)$, es:
      
      $\phi(n)=\phi(p\cdot q)=\phi(p)\cdot\phi(q)=(p-1)\cdot(q-1)$
      
      Porque la función $\phi$ es multiplicativa.

    4.   Hallar la función $\phi$ de Euler de $n$, $\phi(n)$, tal que $\phi(n) = (p-1)\cdot(q-1)$

    5.   Escoge una $e$ tal que $1< e<\phi(n)$ y tal que $e$ y $\phi(n)$ sean coprimos. $e$ es el exponente de la llave publica.
    
    6.   Determinamos $d$, tal que $d\cdot e \equiv 1 \pmod{\phi(n)}$ . Es decir, $d$ es el inverso multiplicativo modular de $e$ modulo $\phi(n)$. La llave privada tiene exponente d y modulo n.

  ***ENCRIPTACION**
    1.  La persona $A$, le transmite su llave publica (de modulo $n$ y exponente $e$) a la persona $B$, manteniendo asi su llave privada en secreto.
    2.  Cuando la persona $B$ quiere enviar el mensaje "M" a la persona $A$, convierte el mensaje M a un entero $m$ tal que $0< m< n $, usando un protocolo reversible acordado, conocido como esquema de relleno.

  ***Definicion**: Los esuqemas de relleno (padding scheme), son métodos que introducen información irrelevante con cierto objetivo. En RSA, generalmente se usa el Optimal Asymmetric Encryption Padding (OAEP), debido a que este esquema añade un elemento de aleatoriedad que puede ser usada para convertir una encriptación determinada en una probabilistica.

    3.   La persona $B$ computa, con la llave publica de la persona $A$, el texto cifrado "c", correspondiente a $c \equiv m^e\pmod n$, es decir, $n|c- m^e$
    4.   La persona $B$ ahora envía el mensaje "M" en texto cifrado $c$, a la persona $A$.
  ***DESCIFRADO**
      1.   La persona $A$ recupera $m$ a partir de $c$ usando su exponente de llave privada, $d$, por $m\equiv c^d \pmod n$
      2.   Dado $m$, la persona $A$ puede recuperar el mensaje original M invirtiendo el esquema de relleno.



3.   Ejemplo: Calculemos el modulo $n$ escogiendo $p = 11$ y $q = 13$, por lo tanto $n= p\cdot q = 11 \cdot 13 = 143$. Y la funcion $\phi$ nos da $\phi(n) = (p-1)(q-1) = 10 \cdot 12 = 120$.

  Para la llave publica escogemos un número primo aleatorio, que sea coprimo de $120$. Escojamos el $7$. Entonces tenemos que $e=7$, y para determinar d, la llave secreta, necesitamos encontrar el inverso multiplicativo de $7 \pmod {120}$. Esto se halla con el algoritmo euclidiano extendido. Entonces tenemos que $d \cdot e \equiv 1 \pmod{\phi(n)}$.
  
  Reemplazando
        *   $d \cdot 7 \equiv 1 \pmod{120}$
        *   $d \equiv \frac{1}{7} \pmod{120}$
        *   $d \equiv 7^{-1} \pmod{120}$
  
  Usando el algoritmo euclidiano:
        *   $120 = 7 \cdot 17 +1$
        *   $1 =120-119$
        *   $1 = 120 - (7\cdot 17)$

  Como estamos trabajando en modulo $120$, podemos tomar el $120$ como $0$ ya que $120 \equiv 0\pmod{120}$
        *   $1 = -7 \cdot 17$
        *   $1 = -17 \cdot 7$
        *   $7^{-1} \pmod{120} \equiv - 17$
  
  Ahora, queremos este numero entre $0$ y $120$, entonces 
        *   $7^{-1} \pmod{120} \equiv 120 - 17$
        *   $7^{-1} \pmod{120} \equiv 103$

  Tenemos nuestro exponente $d$ para la llave privada
        *   $d=103$
  
  Verificamos
        *   $e \cdot d \equiv 1 \pmod{120}$
        *   $7 \cdot 103 = 721 \equiv 1 \pmod{120}$
        *   $721 -1|120$
        *   $720|120$, la cual se cumple puesto que $120 \cdot 6 = 720$
  
  Ahora escogemos nuestro mensaje m es $9$
        *   $c \equiv m^e \pmod n$
        *   $c \equiv 9^7 \pmod {143}$
        *   $c = 48$ 

  Al momento de descifrar, tenemos que  
        *   $m \equiv c^d\pmod n$
        *   $m \equiv 48^{103}\pmod {143}$
        *   $m = 9$ 

  Tenemos lo que probamos por m, mostrando que el algoritmo funciona. Este ejemplo fue muy pequeño comparado con los valores que se usan en las aplicaciones de la vida real.



















```
import math

def verPrimo(n): 
    if (n <= 1):
       return False
    if (n <= 3):
        return True
    if (n % 2 == 0 or n % 3 == 0): 
        return False
    i = 5
    while(i * i <= n): 
        if (n % i == 0 or n % (i + 2) == 0): 
            return False
        i = i + 6
    return True

def toInt(msj):
    msj_i = []
    for character in msj:
        msj_i.append(ord(character))
    return msj_i


def toString(ciphertext):
    char_ciphertext = []
    string = ""
    for number in ciphertext:
        char_ciphertext.append(chr(number))
    return (string.join(char_ciphertext))


def nMax(ciphertext):
    nMaxPos = max(ciphertext)
    return nMaxPos

def encuentraE(phi_n):
    e = 3
    while True:
        if not math.gcd(e,phi_n) == 1:
           e += 2
        else:
            return e

def encuentraD(phi_n, e):
    k = 1
    while True: 
        d, res = divmod(k * phi_n + 1, e)
        if res == 0:
            return d
        k += 1

def encriptacion(m_i, p, q):
    c_i = []
    n = (p * q)
    phi_n = (p - 1) * (q - 1)
    e = encuentraE(phi_n)
    for m in m_i:
        c = pow(m, e, n)
        c_i.append(c) 
    return c_i

def decriptacion(c_i, p, q):
    m_i = []
    n = (p * q)
    phi_n = (p - 1) * (q - 1)
    e = encuentraE(phi_n)
    d = encuentraD(phi_n, e)
    for x in c_i:
        m = pow(x, d, n)
        m_i.append(m)
    return m_i

if __name__ == "__main__":
    msj = input("Escribe el mensaje a encriptar: ")
    while (True):
        print("Elije un numero primo mayor que ", nMax(toInt(msj)))
        p = int(input())
        if (verPrimo(p) == False):
            print(p, "no es primo, intente de nuevo", "\n")
            continue
        else:
            break

    while(True):
        print("Elije otro primo mayor que ", nMax(toInt(msj))," y distinto de ", p)
        q = int(input())
        if (q == p):
            print("Este primo no puede ser igual al anterior, intente de nuevo", "\n")
            continue
        elif (verPrimo(q) == False):
            print(q, "no es primo, intente de nuevo", "\n")
            continue
        else:
            break
    m_i = toInt(msj)
    c_i = encriptacion(m_i, p, q)
    print("Tu mensaje encriptado: ", c_i, "\n")
    print("Si desencriptamos el mensaje ", "\n")
    m_i2 = decriptacion(c_i, p, q)
    descipheredtext = toString(m_i2)
    print("Tu mensaje es: ", descipheredtext)

```
